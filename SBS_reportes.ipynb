### 2. Importamos las librerias necesarias

```python
import pandas as pd
import requests
import os
```

### 3. Eligimos el tipo de empresa del Sistema Financiero a revisar

```python
t_empr = input(
    '''¿Que tipo de empresa desea analizar: 
    0: Banca Múltiple 
    1: Empresas Financieras
    2: Caja Municipal 
    3: Caja Rural
    4: Empresas de Créditos
    
    '''
)
```

### 4. Eligimos el año a descargar

```python
year_start = input(
    ''' ¿Desde qué año se va a consultar:
 (Ejem: 2023)


''')
```

```python
year_end = input(
    ''' ¿Hasta qué año se va a consultar:
 (Ejem: 2024)


''')
```

```python
month = input(
    ''' ¿Hasta qué mes se va a consultar:
 (Ejem: 11, para establecer hasta el mes de noviembre)


''')
```

### 5. Seleccionamos el archivo a descargar

```python
selec_archivo = input(
    ''' Seleccione el archivo a consultar:
    0: Estados Financieros
    1: Req de Patrimonio Efec y Ratio Capital global
    2: Cred Directos según situación

    
    ''')
tipo_empresa_descargar = {'0': {'0':'B-2201','1': 'B-3101', '2':'C-1101', '3':'C-2101', '4':'C-4103'},
                          '1': {'0':'B-2402','1': 'B-3302', '2':'C-1252', '3':'C-2257', '4':'C-4241'}, 
                          '2': {'0':'B-2315','1': 'B-3219', '2':'C-1207', '3':'C-2207', '4':'C-4208'}
                         }
```

### 6. Descargamos los archivos de la SBS

```python
archivo_descargado = {'0': 'EEFF', '1': 'RPF_RCG', '2': 'CD_SITUACION'}
tipo_empresa = {'0': 'Banca Múltiple', '1': 'Empresas Financieras', '2': 'Caja Municipal', '3': 'Caja Rural' , '4': 'Empresas de Créditos' }
ruta_inicial = 'https://intranet2.sbs.gob.pe/estadistica/financiera/'
meses = {0:'Enero',1:'Febrero',2:'Marzo',3:'Abril',4:'Mayo',5:'Junio',6:'Julio',7:'Agosto',8:'Setiembre',9:'Octubre',10:'Noviembre',11:'Diciembre'}
mes_corto = {0:'en', 1:'fe', 2:'ma', 3:'ab', 4:'my', 5:'jn', 6:'jl', 7:'ag', 8:'se', 9:'oc', 10:'no', 11:'di'}

#RUTA DONDE SE GUARDARAN LOS ARCHIVOS
ruta_base = r"C:\Users\lasandoval\Documents\SBS\Estadisticas\BM"

# CREAMOS LA CARPETA DE GUARDADO CON EL NOMBRE DEL AÑO
for year in range(int(year_start),int(year_end) +1):
    year1 = str(year)
    nombre_carpeta = tipo_empresa[t_empr]+'-' + archivo_descargado[selec_archivo] +'_'+year1
    ruta_guardado_anio = os.path.join(ruta_base, nombre_carpeta )
    os.makedirs(ruta_guardado_anio, exist_ok=True)
    if year < int(year_end):
        for m in range(12):
            nombre_archivo = f'{tipo_empresa_descargar[selec_archivo][t_empr]}-{mes_corto[m]}{year1}.XLS'
            url_descarga = f'{ruta_inicial}{year1}/{meses[m]}/{nombre_archivo}'
            # Descargar el archivo
            response = requests.get(url_descarga, stream=True)
            # Verificar si la descarga fue exitosa
            if response.status_code == 200:
                ruta_guardado = os.path.join(ruta_guardado_anio, nombre_archivo)  # Guardar en la carpeta creada
                with open(ruta_guardado, "wb") as file:
                    for chunk in response.iter_content(1024):
                        file.write(chunk)
                print(f"Archivo descargado exitosamente: {ruta_guardado}")
            else:
                print(f"No se pudo descargar el archivo. Código de estado: {response.status_code}")
    else:
        for m in range(int(month)):
            nombre_archivo = f'{tipo_empresa_descargar[selec_archivo][t_empr]}-{mes_corto[m]}{year1}.XLS'
            url_descarga = f'{ruta_inicial}{year1}/{meses[m]}/{nombre_archivo}'
            # Descargar el archivo
            response = requests.get(url_descarga, stream=True)
            # Verificar si la descarga fue exitosa
            if response.status_code == 200:
                ruta_guardado = os.path.join(ruta_guardado_anio, nombre_archivo)  # Guardar en la carpeta creada
                with open(ruta_guardado, "wb") as file:
                    for chunk in response.iter_content(1024):
                        file.write(chunk)
                print(f"Archivo descargado exitosamente: {ruta_guardado}")
            else:
                print(f"No se pudo descargar el archivo. Código de estado: {response.status_code}")
```

### 7. (Opcional) Extraemos los valores de los archivos descargados

```python
#SOLO PARA CRED. DIRECTOS SEGUN SITUACION
pd.set_option('display.float_format', '{:.2f}'.format)
columnas_hoja = {'0':'A:H', '1':'A:G'}
columnas_name = {'0':['Empresas','Vigentes_CP', 'Vigentes_LP','Reestructurados','Refinanciados','Vencidos',
                      'En Cobranza Judicial','Total Creditos Directos'],
                 '1':['Vacio','Empresas','Vigentes_CP', 'Vigentes_LP','Reestructurados','Refinanciados','Vencidos',
                      'En Cobranza Judicial','Total Creditos Directos'],
                 '2':['Empresas','Vigentes_CP', 'Vigentes_LP','Reestructurados y Refinanciados','Vencidos',
                      'En Cobranza Judicial','Total Creditos Directos']}
data = {}
fecha_ef = pd.to_datetime('2015-01-01', format = '%Y-%m-%d')
if int(t_empr) == 1: ##APLICA PARA EMPRESAS FINANCIERAS
    for year in range(int(year_start),int(year_end) +1):
        year1 = str(year)
        for m in range(12):
            fecha = f"{year}{str(m+1).zfill(2)}01"
            fecha1 = pd.to_datetime(fecha, format='%Y%m%d')
            mes1 = fecha1.strftime('%B').capitalize()[:3]
            year2 = fecha1.strftime('%Y')
            nombre_archivo = f'{tipo_empresa_descargar[selec_archivo][t_empr]}-{mes_corto[m]}{year1}.XLS'
            ruta_archivo = f'{ruta_base}\\{tipo_empresa[t_empr]}-{archivo_descargado[selec_archivo]}_{year1}\\{nombre_archivo}'
            # Cargar el archivo
            archivo = pd.read_excel(ruta_archivo)
            try:
                print(f'Archivo {nombre_archivo} Cargado correctamente')
            except:
                print(f'Error al cargar {nombre_archivo}:{e}')
            fecha2 = fecha1.strftime('%m-%Y')
            if fecha1 < fecha_ef:
                archivo.columns = columnas_name['1']
                archivo = archivo.iloc[5:,1:]
            else:
                archivo.columns = columnas_name['0']
                archivo = archivo.iloc[5:,:]
            archivo.reset_index(inplace=True)
            archivo = archivo.iloc[:,1:]
            indice_total = archivo[archivo.iloc[:, 0].str.contains("TOTAL", na=False, case=False)].index.tolist()
            archivo = archivo.iloc[:max(indice_total)+1,:]
            archivo = archivo[~(archivo.Empresas.isna())]
            archivo['FECHA1'] = fecha1
            archivo['FECHA2'] = fecha2
            data[fecha] = archivo

else: ## APLICA PARA LAS DEMAS EMP SISTEMA FINANCIERO
    for year in range(int(year_start),int(year_end) +1):
        year1 = str(year)
        for m in range(12):
            fecha = f"{year}{str(m+1).zfill(2)}01"
            fecha1 = pd.to_datetime(fecha, format='%Y%m%d')
            mes1 = fecha1.strftime('%B').capitalize()[:3]
            year2 = fecha1.strftime('%Y')
            nombre_archivo = f'{tipo_empresa_descargar[selec_archivo][t_empr]}-{mes_corto[m]}{year1}.XLS'
            ruta_archivo = f'{ruta_base}\\{tipo_empresa[t_empr]}-{archivo_descargado[selec_archivo]}_{year1}\\{nombre_archivo}'
            # Cargar el archivo
            archivo = pd.read_excel(ruta_archivo)
            try:
                print(f'Archivo {nombre_archivo} Cargado correctamente')
            except:
                print(f'Error al cargar {nombre_archivo}:{e}')
            fecha2 = fecha1.strftime('%m-%Y')
            if int(t_empr) == 0:
                archivo.columns = columnas_name['0']
            else:
                archivo.columns = columnas_name['2']
            archivo = archivo.iloc[5:,]
            archivo.reset_index(inplace=True)
            archivo = archivo.iloc[:,1:]
            indice_total = archivo[archivo.iloc[:, 0].str.contains("TOTAL", na=False, case=False)].index.tolist()
            archivo = archivo.iloc[:max(indice_total)+1,:]
            archivo = archivo[~(archivo.Empresas.isna())]
            archivo['FECHA1'] = fecha1
            archivo['FECHA2'] = fecha2
            data[fecha] = archivo
data_total = pd.concat(data.values(),axis=0)
data_generada = f'{ruta_base}\\{tipo_empresa[t_empr]}-{archivo_descargado[selec_archivo]}_{year_start}-{year_end}.xlsx'
data_total.to_excel(data_generada, index=False)
try:
    print(f'Archivo {tipo_empresa[t_empr]}-{archivo_descargado[selec_archivo]}_{year_start}-{year_end}.xlsx Descargado correctamente')
except:
    print(f'Error al descargar {tipo_empresa[t_empr]}-{archivo_descargado[selec_archivo]}_{year_start}-{year_end}.xlsx:{e}')
```
